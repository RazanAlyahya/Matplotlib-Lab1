```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe

```python
video=['Game','Comady','Movies']
views=[5000,2000,1000]
plt.bar(video, views)
plt.title('Top Five Video',fontsize = 20)
plt.xlabel('Video',fontsize = 20)
plt.ylabel('Views',fontsize = 20)
plt.grid()
plt.show()
```

```python
data = {'SalesPerson': ['Kathey', 'Michael', 'William',],
        'Region': ['East', 'West', 'North'],
        'OrderAmount': [600, 700, 400],
        'Month': ['Jan', 'Feb', 'Feb'],
       }

SalesDF = pd.DataFrame(data, columns=['SalesPerson', 'Region', 'OrderAmount', 'Month'])
SalesDF  

plt.plot(SalesDF['SalesPerson'],SalesDF['OrderAmount'], c = '0.4')

plt.xlabel('Sales Person')
plt.ylabel('Order Amount')
plt.title('Number of Order Amount every Sales Person')
plt.show()
```

## Q2: Create pie chart with equal explode values.

```python
video=['Game','Comady','Movies']
views=[5000,2000,1000.34]

#add colors
colors = ['#ff9999','#66b3ff','#99ff99']

plt.pie(views,labels=video, autopct='%1.2f%%', colors =colors,  startangle=180, radius=1.5)

plt.show()
```

## Q5: Create three line charts with Legends at left and colors

```python
data = {'video': ['Game','Comady','Movies'],
        'views': [5000,2000,1000],
        'likes': [600, 700, 400],
        'dislikes':[200,300,400],
       }

plt.plot(data['video'],data['views'], c ='green', linestyle = '-', linewidth = 1)
plt.plot(data['video'],data['likes'], c ='red', linestyle = '--', linewidth = 2)
plt.plot(data['video'],data['dislikes'], c ='b', linestyle = 'solid', linewidth = 3)
plt.legend(['views','likes','dislikes'], shadow = True, loc = 0) # location (0 = best)
plt.ylabel('number')
plt.xlabel('video')
plt.title('Number of video')
plt.show()
```
